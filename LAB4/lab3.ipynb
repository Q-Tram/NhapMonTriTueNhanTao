## Lý thuyết
Thuật toán di truyền (Genetic Algorithm - GA) là một phương pháp tìm kiếm dựa trên cơ chế tiến hóa tự nhiên, lấy cảm hứng từ thuyết tiến hóa của Darwin. GA được sử dụng để tìm lời giải gần tối ưu cho các bài toán tối ưu hóa phức tạp.

### Các bước chính của thuật toán:
1. **Khởi tạo quần thể**: Tạo ngẫu nhiên một tập hợp các cá thể (giải pháp tiềm năng).
2. **Đánh giá độ thích nghi (Fitness Evaluation)**: Tính giá trị hàm mục tiêu cho mỗi cá thể.
3. **Lựa chọn (Selection)**: Chọn các cá thể tốt nhất để lai ghép (thường sử dụng các phương pháp như Tournament Selection hoặc Roulette Wheel).
4. **Lai ghép (Crossover)**: Kết hợp hai cá thể để tạo ra cá thể mới.
5. **Đột biến (Mutation)**: Thay đổi ngẫu nhiên một số gen để duy trì tính đa dạng.
6. **Lặp lại**: Quay lại bước 2 cho đến khi đạt điều kiện dừng (số thế hệ hoặc giá trị fitness mong muốn).

## Ví dụ 1: Tối ưu hóa hàm một biến

### Bài toán  
Tìm $x$ sao cho  
$$
f(x) = -(x^2) + 10x + 50
$$  
đạt giá trị lớn nhất trong khoảng  
$$
x \in [0, 10].
$$

# Mã giả
```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection:
        - Chọn ngẫu nhiên tournament_size cá thể trong quần thể
        - Chọn cá thể có fitness cao nhất làm cha mẹ 1
    Chọn cha mẹ 2 bằng Tournament Selection tương tự
    
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới có giá trị trung bình của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
            
    Thực hiện đột biến cho con mới với xác suất mutation_rate:
        Nếu đột biến:
            Gán con mới giá trị ngẫu nhiên trong [min_val, max_val]
            
    Thêm con mới vào quần thể mới
    
Cập nhật quần thể hiện tại = quần thể mới
In ra cá thể tốt nhất và giá trị fitness tương ứng của thế hệ


````python
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return -(x**2) + 10*x + 50

# Khởi tạo quần thể
def initialize_population(pop_size, min_val, max_val):
    return [random.uniform(min_val, max_val) for _ in range(pop_size)]

# Lựa chọn (Tournament Selection)
def select_parents(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return (parent1 + parent2) / 2
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.1, min_val=0, max_val=10):
    if random.random() < mutation_rate:
        return random.uniform(min_val, max_val)
    return individual

# Thuật toán di truyền
def genetic_algorithm_example1(pop_size=50, generations=100, min_val=0, max_val=10):
    population = initialize_population(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function(x) for x in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents(population, fitness_values)
            parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function(x) for x in population])]
        print(f'Thế hệ {generation + 1}: x = {best_individual:.4f}, f(x) = {best_fitness:.4f}')
    
    best_fitness = max([fitness_function(x) for x in population])
    best_individual = population[np.argmax([fitness_function(x) for x in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị f(x) tốt nhất')
    plt.title('Sự thay đổi của f(x) qua các thế hệ')
    plt.show()
    
    return best_individual, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm_example1()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, f(x) = {best_f:.4f}')
````

### Mã giả 

```plaintext
Tạo quần thể mới rỗng
Lặp cho đến khi quần thể mới đủ N cá thể:
    Chọn cha mẹ 1 bằng Tournament Selection trên quần thể hiện tại
    Chọn cha mẹ 2 bằng Tournament Selection trên quần thể hiện tại
    Thực hiện lai ghép (crossover) giữa cha mẹ 1 và cha mẹ 2 với xác suất crossover_rate:
        Nếu crossover thành công:
            Tạo con mới bằng trung bình giá trị (x, y) của cha mẹ
        Ngược lại:
            Con mới là cha mẹ 1
    Thực hiện đột biến cho con mới với xác suất mutation_rate từng biến:
        Nếu đột biến:
            Gán giá trị ngẫu nhiên trong [min_val, max_val] cho biến tương ứng
    Thêm con mới vào quần thể mới

Cập nhật quần thể hiện tại = quần thể mới
In kết quả tốt nhất thế hệ

````python
# Hàm mục tiêu
def fitness_function_2d(x, y):
    return -(x**2 + y**2)  # Sử dụng dấu âm vì GA tìm max

# Khởi tạo quần thể
def initialize_population_2d(pop_size, min_val, max_val):
    return [(random.uniform(min_val, max_val), random.uniform(min_val, max_val)) for _ in range(pop_size)]

# Lựa chọn
def select_parents_2d(population, fitness_values, tournament_size=3):
    selected = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected, key=lambda i: fitness_values[i])
    return population[best_idx]

# Lai ghép
def crossover_2d(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        return ((parent1[0] + parent2[0]) / 2, (parent1[1] + parent2[1]) / 2)
    return parent1

# Đột biến
def mutate_2d(individual, mutation_rate=0.1, min_val=-5, max_val=5):
    x, y = individual
    if random.random() < mutation_rate:
        x = random.uniform(min_val, max_val)
    if random.random() < mutation_rate:
        y = random.uniform(min_val, max_val)
    return (x, y)

# Thuật toán di truyền
def genetic_algorithm_example2(pop_size=50, generations=100, min_val=-5, max_val=5):
    population = initialize_population_2d(pop_size, min_val, max_val)
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = [fitness_function_2d(x, y) for x, y in population]
        best_fitness = max(fitness_values)
        best_fitness_history.append(-best_fitness)  # Chuyển về giá trị thực
        
        new_population = []
        for _ in range(pop_size):
            parent1 = select_parents_2d(population, fitness_values)
            parent2 = select_parents_2d(population, fitness_values)
            child = crossover_2d(parent1, parent2)
            child = mutate_2d(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
        print(f'Thế hệ {generation + 1}: (x, y) = ({best_individual[0]:.4f}, {best_individual[1]:.4f}), g(x, y) = {-best_fitness:.4f}')
    
    best_fitness = max([fitness_function_2d(x, y) for x, y in population])
    best_individual = population[np.argmax([fitness_function_2d(x, y) for x, y in population])]
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị g(x, y) tốt nhất')
    plt.title('Sự thay đổi của g(x, y) qua các thế hệ')
    plt.show()
    
    return best_individual, -best_fitness

# Chạy thuật toán
best_xy, best_g = genetic_algorithm_example2()
print(f'\nKết quả cuối cùng: (x, y) = ({best_xy[0]:.4f}, {best_xy[1]:.4f}), g(x, y) = {best_g:.4f}')
````

# Bài tập 1
# Thay đổi hàm mục tiêu**: Thử áp dụng thuật toán cho hàm $h(x) = sin(x) + cos(x)$ trong khoảng $x \in [0, 2\pi]$. So sánh kết quả với nghiệm chính xác.

## Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

## Mục tiêu
- Tối ưu hàm:
  $$
  h(x) = \sin(x) + \cos(x)
  $$
  trong miền:
  $$
  x \in [0, 2\pi]
  $$
- Tìm giá trị **x** sao cho **h(x)** đạt giá trị lớn nhất.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

## Ý tưởng thuật toán di truyền (GA)

1. **Mã hóa**: Biểu diễn giá trị \( x \) bằng chuỗi nhị phân dài 16 bit.  
2. **Giải mã**: Chuyển từ chuỗi nhị phân sang giá trị thực trong khoảng \([0, 2\pi]\).  
3. **Fitness**: Sử dụng chính giá trị \( h(x) \) làm độ thích nghi.  
4. **Chọn lọc**: Roulette Wheel Selection.  
5. **Lai ghép**: Lai 1 điểm với xác suất cao.  
6. **Đột biến**: Lật bit với xác suất thấp.

---


````python
# code vào  đây nhe
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def fitness_function(x):
    return np.sin(x) + np.cos(x)

# Mã hóa: Chuyển đổi giá trị thực thành chuỗi nhị phân
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Giải mã: Chuyển đổi chuỗi nhị phân thành giá trị thực
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tính độ thích nghi
def calculate_fitness(population):
    fitness_values = [fitness_function(decode(individual)) for individual in population]
    # Đảm bảo không có giá trị âm
    return [max(f, 0) for f in fitness_values]

# Chọn lọc (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness == 0:  # Tránh chia cho 0
        return random.choices(population, k=2)  # Chọn ngẫu nhiên nếu không có độ thích nghi

    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Lai ghép
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        crossover_point = random.randint(1, len(parent1) - 1)
        child = parent1[:crossover_point] + parent2[crossover_point:]
        return child
    return parent1

# Đột biến
def mutate(individual, mutation_rate=0.01):
    mutated = ''.join(bit if random.random() > mutation_rate else '1' if bit == '0' else '0' for bit in individual)
    return mutated

# Thuật toán di truyền
def genetic_algorithm(pop_size=50, generations=100, bits=16, min_val=0, max_val=2 * np.pi):
    # Khởi tạo quần thể
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for generation in range(generations):
        fitness_values = calculate_fitness(population)
        best_fitness = max(fitness_values)
        best_fitness_history.append(best_fitness)
        
        new_population = []
        for _ in range(pop_size):
            parent1, parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        best_individual = population[np.argmax(fitness_values)]
        best_x = decode(best_individual)
        print(f'Thế hệ {generation + 1}: x = {best_x:.4f}, h(x) = {best_fitness:.4f}')
    
    best_fitness = max(calculate_fitness(population))
    best_individual = population[np.argmax(calculate_fitness(population))]
    best_x = decode(best_individual)
    
    # Trực quan hóa
    plt.plot(range(generations), best_fitness_history)
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tốt nhất')
    plt.title('Sự thay đổi của h(x) qua các thế hệ')
    plt.show()
    
    return best_x, best_fitness

# Chạy thuật toán
best_x, best_f = genetic_algorithm()
print(f'\nKết quả cuối cùng: x = {best_x:.4f}, h(x) = {best_f:.4f}')
````

# Bài tập 2: Tối ưu hàm $h(x) = \sin(x) + \cos(x)$ bằng Thuật Toán Di Truyền (Genetic Algorithm)

---

## Mục tiêu

Tối ưu hàm:
$$
h(x) = \sin(x) + \cos(x)
$$
trong khoảng:
$$
x \in [0, 2\pi]
$$

> Tìm giá trị $x$ sao cho $h(x)$ đạt **giá trị lớn nhất**.

---

## Phân tích nghiệm chính xác

Ta có:
$$
h(x) = \sin(x) + \cos(x) = \sqrt{2} \cdot \sin\left(x + \frac{\pi}{4}\right)
\Rightarrow \max h(x) = \sqrt{2} \approx 1.4142
$$

Xảy ra khi:
$$
x = \frac{\pi}{4} \approx 0.7854
$$

---

##  Ý tưởng thuật toán di truyền (GA)

| Thành phần        | Chi tiết                                                                 |
|------------------|-------------------------------------------------------------------------|
| **Mã hóa**        | Dùng chuỗi nhị phân 16 bit để biểu diễn giá trị $x \in [0, 2\pi]$         |
| **Giải mã**       | Ánh xạ nhị phân sang giá trị thực: $x = 0 + \dfrac{\text{decimal}}{2^{16}-1} \cdot 2\pi$ |
| **Fitness**       | Chính là $h(x) = \sin(x) + \cos(x)$                                     |
| **Chọn lọc**      | Roulette Wheel Selection                                                |
| **Lai ghép**      | 1 điểm cắt với xác suất 0.8                                              |
| **Đột biến**      | Xác suất 5% (0.05)                                                       |

---

````python
# code vào đây nhe
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def target_function(x):
    return np.sin(x) + np.cos(x)

# Tính fitness - Đảm bảo luôn dương bằng cách cộng thêm 2 (vì min của sin(x)+cos(x) là -√2 ≈ -1.414)
def fitness_function(x):
    return target_function(x) + 2  # Dịch lên để luôn >= 0

# Mã hóa: Chuyển đổi giá trị thực thành chuỗi nhị phân
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Giải mã: Chuỗi nhị phân -> giá trị thực
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tính toán fitness cho quần thể
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chọn lọc (Roulette Wheel Selection) - Đã được sửa để xử lý trường hợp đặc biệt
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    
    # Xử lý trường hợp đặc biệt khi tổng fitness = 0 hoặc có giá trị âm
    if total_fitness <= 0:
        return random.choices(population, k=2)
    
    selection_probs = [f/total_fitness for f in fitness_values]
    try:
        parents = np.random.choice(population, size=2, p=selection_probs)
        return parents[0], parents[1]
    except ValueError:
        return random.choices(population, k=2)

# Lai ghép 1 điểm với xác suất 0.8
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1)-1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# Đột biến với xác suất 5% (0.05)
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thuật toán di truyền chính
def genetic_algorithm(pop_size=50, generations=100, bits=16, min_val=0, max_val=2*np.pi):
    # Khởi tạo quần thể ngẫu nhiên
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for gen in range(generations):
        # Tính fitness
        fitness_values = calculate_fitness(population)
        target_values = [target_function(decode(ind)) for ind in population]
        best_target = max(target_values)
        best_fitness_history.append(best_target)
        
        # Tạo quần thể mới
        new_population = []
        for _ in range(pop_size):
            # Chọn lọc
            parent1, parent2 = select_parents(population, fitness_values)
            # Lai ghép
            child = crossover(parent1, parent2)
            # Đột biến
            child = mutate(child)
            new_population.append(child)
        
        population = new_population
        
        # Hiển thị kết quả tốt nhất mỗi thế hệ
        best_idx = np.argmax(target_values)
        best_x = decode(population[best_idx])
        print(f"Gen {gen+1}: x = {best_x:.4f}, h(x) = {best_target:.4f}")
    
    # Kết quả cuối cùng
    final_fitness = calculate_fitness(population)
    final_target = [target_function(decode(ind)) for ind in population]
    best_idx = np.argmax(final_target)
    best_x = decode(population[best_idx])
    best_h = final_target[best_idx]
    
    # Vẽ đồ thị
    plt.plot(range(generations), best_fitness_history)
    plt.axhline(y=np.sqrt(2), color='r', linestyle='--', label='Max theoretical')
    plt.xlabel('Generation')
    plt.ylabel('Best h(x)')
    plt.title('Genetic Algorithm Performance')
    plt.legend()
    plt.show()
    
    return best_x, best_h

# Chạy thuật toán
best_x, best_h = genetic_algorithm()
print(f"\nThế hệ: x = {best_x:.4f}, h(x) = {best_h:.4f}")
print(f"Kết quả cuối cùng: x = {np.pi/4:.4f}, h(x) = {np.sqrt(2):.4f}")
````

## Phân tích ảnh hưởng

- **`pop_size`**: Kích thước quần thể lớn giúp tìm kiếm toàn cục tốt hơn nhưng chậm hơn, kích thước nhỏ làm thuật toán hội tụ nhanh nhưng dễ mắc kẹt.
- **`generations`**: Số thế hệ lớn tăng khả năng cải tiến kết quả nhưng làm tăng thời gian tính toán.
- **`mutation_rate`**: Tỷ lệ đột biến vừa phải giúp duy trì đa dạng di truyền, tránh kẹt cục bộ; quá cao hoặc quá thấp đều làm giảm hiệu quả.
- **`crossover_rate`**: Tỷ lệ lai ghép hợp lý giúp khai thác tốt thông tin di truyền, tăng hiệu quả hội tụ.

````python
# code vào đây 
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def target_function(x):
    return np.sin(x) + np.cos(x)

# Tính fitness - Đảm bảo luôn dương bằng cách cộng thêm 2
def fitness_function(x):
    return target_function(x) + 2

# Mã hóa: Chuyển đổi giá trị thực thành chuỗi nhị phân
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Giải mã: Chuỗi nhị phân -> giá trị thực
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tính toán fitness cho quần thể
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chọn lọc (Roulette Wheel Selection)
def select_parents(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness <= 0:
        return random.choices(population, k=2)
    
    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Lai ghép 1 điểm với xác suất crossover_rate
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# Đột biến với xác suất mutation_rate
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thuật toán di truyền chính
def genetic_algorithm(pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, bits=16, min_val=0, max_val=2*np.pi):
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for gen in range(generations):
        fitness_values = calculate_fitness(population)
        best_target = max(fitness_values)
        best_fitness_history.append(best_target)
        
        new_population = []
        for _ in range(pop_size):
            parent1, parent2 = select_parents(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        
        population = new_population
        
        # Hiển thị kết quả tốt nhất mỗi thế hệ
        best_idx = np.argmax(fitness_values)
        best_x = decode(population[best_idx])
        print(f"Thế hệ {gen + 1}: x = {best_x:.4f}, h(x) = {best_target:.4f}")
    
    return best_fitness_history

# Phân tích ảnh hưởng của các tham số
def analyze_parameters():
    # Thay đổi kích thước quần thể
    pop_sizes = [10, 50, 100]
    generations = 100
    mutation_rate = 0.05
    crossover_rate = 0.8
    
    for pop_size in pop_sizes:
        print(f"\nKích thước quần thể: {pop_size}")
        fitness_history = genetic_algorithm(pop_size=pop_size, generations=generations, mutation_rate=mutation_rate, crossover_rate=crossover_rate)
        plt.plot(fitness_history, label=f'pop_size={pop_size}')
    
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tốt nhất')
    plt.title('Ảnh hưởng của kích thước quần thể đến tốc độ hội tụ')
    plt.legend()
    plt.show()

    # Thay đổi tỷ lệ đột biến
    mutation_rates = [0.01, 0.05, 0.1]
    
    for mutation_rate in mutation_rates:
        print(f"\nTỷ lệ đột biến: {mutation_rate}")
        fitness_history = genetic_algorithm(pop_size=50, generations=generations, mutation_rate=mutation_rate, crossover_rate=crossover_rate)
        plt.plot(fitness_history, label=f'mutation_rate={mutation_rate}')
    
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tốt nhất')
    plt.title('Ảnh hưởng của tỷ lệ đột biến đến tốc độ hội tụ')
    plt.legend()
    plt.show()

    # Thay đổi tỷ lệ lai ghép
    crossover_rates = [0.6, 0.8, 1.0]
    
    for crossover_rate in crossover_rates:
        print(f"\nTỷ lệ lai ghép: {crossover_rate}")
        fitness_history = genetic_algorithm(pop_size=50, generations=generations, mutation_rate=mutation_rate, crossover_rate=crossover_rate)
        plt.plot(fitness_history, label=f'crossover_rate={crossover_rate}')
    
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tốt nhất')
    plt.title('Ảnh hưởng của tỷ lệ lai ghép đến tốc độ hội tụ')
    plt.legend()
    plt.show()

# Chạy phân tích
analyze_parameters()
````

````python
# code vào đây nhé
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def target_function(x):
    return np.sin(x) + np.cos(x)

# Tính fitness - Đảm bảo luôn dương bằng cách cộng thêm 2
def fitness_function(x):
    return target_function(x) + 2

# Mã hóa: Chuyển đổi giá trị thực thành chuỗi nhị phân
def encode(x, bits=16, min_val=0, max_val=2 * np.pi):
    normalized_value = (x - min_val) / (max_val - min_val)
    integer_value = int(normalized_value * (2**bits - 1))
    return format(integer_value, f'0{bits}b')

# Giải mã: Chuỗi nhị phân -> giá trị thực
def decode(binary_string, bits=16, min_val=0, max_val=2 * np.pi):
    integer_value = int(binary_string, 2)
    normalized_value = integer_value / (2**bits - 1)
    return min_val + normalized_value * (max_val - min_val)

# Tính toán fitness cho quần thể
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chọn lọc ngẫu nhiên
def random_selection(population):
    return random.choice(population), random.choice(population)

# Chọn lọc (Roulette Wheel Selection)
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness <= 0:
        return random_selection(population)
    
    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Chọn lọc (Tournament Selection)
def tournament_selection(population, fitness_values, tournament_size=3):
    selected_indices = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected_indices, key=lambda i: fitness_values[i])
    return population[best_idx], population[best_idx]

# Lai ghép 1 điểm với xác suất crossover_rate
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# Đột biến với xác suất mutation_rate
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thuật toán di truyền chính
def genetic_algorithm(selection_method, pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, bits=16, min_val=0, max_val=2*np.pi):
    population = [encode(random.uniform(min_val, max_val), bits) for _ in range(pop_size)]
    best_fitness_history = []
    
    for gen in range(generations):
        fitness_values = calculate_fitness(population)
        best_target = max(fitness_values)
        best_fitness_history.append(best_target)
        
        new_population = []
        for _ in range(pop_size):
            if selection_method == 'random':
                parent1, parent2 = random_selection(population)
            elif selection_method == 'roulette':
                parent1, parent2 = roulette_wheel_selection(population, fitness_values)
            elif selection_method == 'tournament':
                parent1, parent2 = tournament_selection(population, fitness_values)
            else:
                raise ValueError("Unknown selection method")
            
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        
        population = new_population
        
        # Hiển thị kết quả tốt nhất mỗi thế hệ
        best_idx = np.argmax(fitness_values)
        best_x = decode(population[best_idx])
        print(f"Gen {gen + 1}: x = {best_x:.4f}, h(x) = {best_target:.4f}")
    
    return best_fitness_history

# Phân tích hiệu quả của các phương pháp lựa chọn
def compare_selection_methods():
    methods = ['random', 'roulette', 'tournament']
    results = {}
    
    for method in methods:
        print(f"\nChạy với phương pháp lựa chọn: {method}")
        fitness_history = genetic_algorithm(selection_method=method, pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8)
        results[method] = fitness_history
        plt.plot(fitness_history, label=f'Selection: {method}')
    
    plt.xlabel('Thế hệ')
    plt.ylabel('Giá trị h(x) tốt nhất')
    plt.title('So sánh các phương pháp lựa chọn trong thuật toán di truyền')
    plt.legend()
    plt.show()

# Chạy phân tích
compare_selection_methods()
````

# Trực quan hóa quần thể – Vẽ biểu đồ phân tán (scatter plot) thể hiện sự phân bố của quần thể qua các thế hệ

## Ý tưởng thực hiện

- Nếu không thể trực quan 3D trực tiếp, có thể vẽ scatter 2D từng cặp biến: $(x, y)$, $(y, z)$, $(x, z)$ qua các thế hệ.

---

````python
#code vào đây
import numpy as np
import random
import matplotlib.pyplot as plt

# Hàm mục tiêu
def target_function(variables):
    return sum(x**2 for x in variables)

# Tính fitness - Đảm bảo luôn dương
def fitness_function(variables):
    return -target_function(variables)  # Tối ưu hóa hàm k(x,y,z)

# Mã hóa: Chuyển đổi giá trị thực thành chuỗi nhị phân
def encode(variables, bits=16, min_val=-10, max_val=10):
    encoded = []
    for var in variables:
        normalized_value = (var - min_val) / (max_val - min_val)
        integer_value = int(normalized_value * (2**bits - 1))
        encoded.append(format(integer_value, f'0{bits}b'))
    return ''.join(encoded)  # Trả về chuỗi nhị phân duy nhất

# Giải mã: Chuỗi nhị phân -> giá trị thực
def decode(encoded_str, bits=16, min_val=-10, max_val=10):
    decoded = []
    for i in range(0, len(encoded_str), bits):
        binary_string = encoded_str[i:i + bits]
        integer_value = int(binary_string, 2)
        normalized_value = integer_value / (2**bits - 1)
        decoded.append(min_val + normalized_value * (max_val - min_val))
    return decoded

# Tính toán fitness cho quần thể
def calculate_fitness(population):
    return [fitness_function(decode(ind)) for ind in population]

# Chọn lọc (Roulette Wheel Selection)
def roulette_wheel_selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    if total_fitness <= 0:
        return random.choice(population), random.choice(population)
    
    selection_probs = [f / total_fitness for f in fitness_values]
    return np.random.choice(population, size=2, p=selection_probs)

# Lai ghép 1 điểm với xác suất crossover_rate
def crossover(parent1, parent2, crossover_rate=0.8):
    if random.random() < crossover_rate:
        point = random.randint(1, len(parent1) - 1)
        child = parent1[:point] + parent2[point:]
        return child
    return parent1

# Đột biến với xác suất mutation_rate
def mutate(individual, mutation_rate=0.05):
    return ''.join(
        bit if random.random() > mutation_rate 
        else '1' if bit == '0' else '0' 
        for bit in individual
    )

# Thuật toán di truyền chính
def genetic_algorithm(pop_size=50, generations=100, mutation_rate=0.05, crossover_rate=0.8, bits=16, min_val=-10, max_val=10):
    population = [encode([random.uniform(min_val, max_val) for _ in range(3)], bits, min_val, max_val) for _ in range(pop_size)]
    best_fitness_history = []
    all_population_history = []  # Lưu trữ quần thể qua các thế hệ
    
    for gen in range(generations):
        fitness_values = calculate_fitness(population)
        best_target = max(fitness_values)
        best_fitness_history.append(best_target)
        
        # Lưu trữ quần thể hiện tại
        all_population_history.append([decode(ind) for ind in population])
        
        new_population = []
        for _ in range(pop_size):
            parent1, parent2 = roulette_wheel_selection(population, fitness_values)
            child = crossover(parent1, parent2, crossover_rate)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        
        population = new_population
        
        # Hiển thị kết quả tốt nhất mỗi thế hệ
        best_idx = np.argmax(fitness_values)
        best_vars = decode(population[best_idx])
        print(f"Gen {gen + 1}: x = {best_vars[0]:.4f}, y = {best_vars[1]:.4f}, z = {best_vars[2]:.4f}, fitness = {best_target:.4f}")
    
    return best_fitness_history, all_population_history

# Trực quan hóa quần thể
def visualize_population(all_population_history, step=10):
    for gen in range(0, len(all_population_history), step):
        population = np.array(all_population_history[gen])
        
        # Vẽ scatter plot cho từng cặp biến
        plt.figure(figsize=(12, 8))
        plt.scatter(population[:, 0], population[:, 1], label='(x, y)', alpha=0.5)
        plt.scatter(population[:, 1], population[:, 2], label='(y, z)', alpha=0.5)
        plt.scatter(population[:, 0], population[:, 2], label='(x, z)', alpha=0.5)
        plt.title(f'Population Distribution at Generation {gen + 1}')
        plt.xlabel('Variable 1')
        plt.ylabel('Variable 2')
        plt.legend()
        plt.grid()
        plt.show()

# Chạy thuật toán và trực quan hóa
best_fitness_history, all_population_history = genetic_algorithm()
visualize_population(all_population_history)
````
